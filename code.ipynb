```python
text_columns = ['Food_Name', 'Category', 'Veg_NonVeg', 'Target_Mood', 'Description']

for col in text_columns:
    df[col] = df[col].str.strip()
```

```python
df['Category'] = df['Category'].str.title()
df['Veg_NonVeg'] = df['Veg_NonVeg'].str.title()
df['Target_Mood'] = df['Target_Mood'].str.title()
```

```python
df['Calories'] = df['Calories'].astype(int)
df['Cooking_Time_Min'] = df['Cooking_Time_Min'].astype(int)
df['Rating'] = df['Rating'].astype(float)
```

```python
# Check for Negative Values
df = df[df['Calories'] >= 0]
df = df[df['Cooking_Time_Min'] >= 0]
```

```python
df = df.dropna()
```

```python
df.reset_index(drop=True, inplace=True)
```
